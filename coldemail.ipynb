```python
from langchain_community.document_loaders import WebBaseLoader
loader = WebBaseLoader("https://jobs.nike.com/job/R-37999")
page_data = loader.load().pop().page_content
print(page_data)
```

```python
from langchain_core.prompts import PromptTemplate

prompt_template = PromptTemplate.from_template(
    """
    ## SCRAPED TEXT FROM WEBSITE:
    {page_data}

    The scraped text is from the career's page of a website.

    ### INSTRUCTION:

    Your job is to extract the job postings and return them in JSON format containing

    following keys: 'role', 'experience', 'skills', and 'description'.

    Only return the valid JSON.

    ### VALID JSON (NO PREAMBLE):
    """
)

chain_extract = prompt_template | llm

res = chain_extract.invoke(input={'page_data': page_data})
print(res.content)
```

```python
type(res.content)
```

```python
from langchain_core.output_parsers import JsonOutputParser
```

```python
json_parser = JsonOutputParser()
json_res = json_parser.parse(res.content)
print(json_res)
```

```python
import chromadb
import uuid
import pandas as pd

# Initialize the Persistent Client
client = chromadb.PersistentClient('vectorstore')

# Get or create the collection
collection = client.get_or_create_collection(name="portfolio")

# Check if the collection is empty
if not collection.count():
    # Assuming df is your DataFrame containing the tech stack and links
    for _, row in df.iterrows():
        try:
            # Add document to the collection
            collection.add(
                documents=row["Techstack"],  # Document content
                metadatas={"links": row["Links"]},  # Metadata with links
                ids=[str(uuid.uuid4())]  # Generate a unique ID
            )
            print(f"Added document for tech stack: {row['Techstack']}")
        except Exception as e:
            print(f"Error adding document: {e}")
else:
    print("Collection already has documents.")
```

```python
links = collection.query(query_texts=["Experience in python" , "Expertise in React Native" ] , n_results=2).get('metadatas' , [])
```
